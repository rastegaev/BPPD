```python
###############################################################################################################
################################## Типы данных Python
###############################################################################################################

# Тип данных "Строка"
name = "Ivan"       # строка "Ivan"
print(type(name))   # на экране будет <class 'str'>

# Тип данных "целое число"
n = 5               # целое число
print(type(n))      # <class 'int'>

# Тип данных "с плавающей точкой" (floating point) или как у нас было принято "дробные числа"
r = 5.0             # вещественное число
print(type(r))      # <class 'float'>

# Тип данных "Логический" True (правда) или False (ложь), также True = 1, False = 0
x = True            # логический тип
print(type(x))      # <class 'bool'>
```

```python
###############################################################################################################
################################## Работа с простыми числами
###############################################################################################################

# Импортируем библиотеку работы с математическими функциями, библиотека входит в стандартный набор
import math

x = 5
y = 10
z = 2
neg = -5

print ("x =", x, "y =", y, "z =", z, "neg =", neg)
print ("x + y - z =", (x + y - z) )
print ("(x * z) / z =", ((x * z) / z) )
print ("(возведение в степень вар 1) x^z =", x**z)
print ("(возведение в степень вар 2) x^z =", pow(x, z))

# Оператор остатка от деления возвращает остаток от деления одного числа на другое
print ("Деление без остатка (выводит остаток от деления) x % y =", x % y, ", 5 % 5 =", 5 % 5, ", 10 % 100 =", 10 % 100)

# Оператор целочисленного деления округляет результат до целого числа не большего результата (к ближайшему меньшему целому)
print ("Целочисленное деление (округление до целого) x // y =", x // y, ", 5 // 5 =", 5 // 5, ", 10 // 100 =", 10 // 100, 
       ", 327 // 23 =", 327 // 23)
# Поиск модуля числа
print ("Модуль числа: |neg| =", abs(neg), ", |10| =", abs(10), ", |-10| =", abs(-10))

# Нахождение квадратного корня из числа
print ("Квадратный корень: √y =", math.sqrt(y), ", √25 =", math.sqrt(25), ", √33 =", math.sqrt(33), ", √|-25| =", math.sqrt(abs(-25)) )
```

```python
###############################################################################################################
################################### Логический тип данных / Булевы значения (Булева алгебра)
###############################################################################################################

a = True
b = False
c = True
# Операция логического И
print ("Операция логического И (вариант 1):")
print ("a И b =", (a and b)) # чтобы получить True - необходимо, чтобы и а и b были True
print ("a И c =", (a and c)) # тут мы видим, что и а и с оба True, соответственно получаем True
# Другой вариант И (&)
print ("Операция логического И (вариант 2):")
print ("a И b =", (a & b)) # чтобы получить True - необходимо, чтобы и а и b были True
print ("a И c =", (a & c)) # тут мы видим, что и а и с оба True, соответственно получаем True

# Операция NOT (логическое отрицание)
print ("Операция логического отрицания:")
print ("НЕ a =", (not a)) # инверсия значения, если было True, то станет False
print ("НЕ b =", (not b)) # инверсия значения, если было False, то станет True

# Операция логического ИЛИ
print ("Операция логического отрицания:")
print ("а ИЛИ b =", (a or b)) # если хотя бы одно значения является True, то и весь результат будет True
print ("a ИЛИ c =", (a or c)) # False будет только тогда, когда обе переменные будут False

# Сложение
print ("Операция сложения/вычитания логических типов (True = 1 False = 0):")
print ("а + b =", (a + b)) # 
print ("a - c =", (a - c)) # 
print ("a + c =", (a + c)) # 

# Сравнение
print ("Операция сравнения логических типов (True = 1 False = 0):")
print ("(Больше?) а > b =", (a > b)) # 
print ("(Меньше?) a < c =", (a < c)) # 
print ("(Равно(1)?) a = c =", (a == c)) # 
print ("(Равно(2)?) a = c =", (a is c)) # 
print ("(Больше или равно?) a >= c =", (a >= c)) # 
print ("(Меньше или равно?) a <= c =", (a <= c)) # 
print ("(Не равно(1)?) a ≠ c =", (a != c)) # 
print ("(Не равно(2)?) a ≠ c =", (a is not c)) # 
```

```python
###############################################################################################################
################################### Строки (Strings)
###############################################################################################################
myString = "python - великий язык программирования."
print("Просто выводим нашу строковую переменную:", myString)
print(type(myString)) # https://www.w3schools.com/python/python_ref_string.asp
print("А теперь ее же, но первую букву сделаем заглавной:", myString.capitalize())
# Присвоим результат выполнения функции(метода) в новую переменную
myUpperString = myString.upper()
print("Давай проверим в каком регистре у нас наша новая строка myUpperString")
print("Неужели в в нижнем регистре?", myUpperString.islower())
print("Или все же в верхнем?", myUpperString.isupper())
print("НАПИШЕМ ВСЕ КАПСОМ!!!:", myUpperString)
print("Немного сбавим обороты и успокоимся:", myUpperString.lower())
print("Посмотрим является ли наша строка идентификатором?", myUpperString.isidentifier())
print("А может быть это число?", myUpperString.isdigit())
print("Так, окей, а может быть строка '65' будет числом?", "65".isdigit())
print("Является ли наша строка пробельным символом?", myUpperString.isspace())
# Давай создадим строку, которая будет состоять только из пробелов?
whiteSpaceString = "     "
print("Проверим состоит ли только что созданная строка из пробелов?", whiteSpaceString.isspace())
# А теперь представь, что ты разработчик сайта и делаешь форму ввода пароля для пользователя
# Пользователь не видит, что он вводит в твое поле ввода, и он случайно вводит пароль, где в конце или в начале может затесаться пробел
# Создадим новую переменную строкового типа
password = "  mYCooLPassWord      " # Как ты видишь, тут дохера пробелов, нужно бы убрать их, так?
print("Уберем все пробелы справа из строки:", password.rstrip())
print("Уберем все пробелы слева из строки:", password.lstrip())
print("Уберем все пробелы слева и справа из строки:", password.strip())
print("        ")
# Теперь мы создадим еще одну новую строку, по которой будем проводить анализ кала
some_strange_string_variable="""Жрать! Пожрать от пуза всяких ништяков ; shit lol kek cheburek ; lol HACHAPURI S SIROM!!! ;
 lol SHAWARMA U METRO; shit shit shit, ДИКО СРАТЬ!!!!"""
print ("Выведем наш небольшой план на день:", some_strange_string_variable)
print ("Давай разобьем наш план на этапы через ;")
planSteps = some_strange_string_variable.split(";") # split позволяет разбить строку на массив строк через определенный разделитель, в нашем случае - ;
print(type(planSteps)) # https://www.w3schools.com/python/python_ref_list.asp
print ("Сколько шагов в нашем плане?", len(planSteps))
# В массивах индекс первого элемента начинается с 0
print ("Шаг 1", planSteps[0])
print ("Шаг 2", planSteps[1])
print ("Шаг 3", planSteps[2])
print ("Шаг 4", planSteps[3])
print ("Шаг 5", planSteps[4])
# У нас получилась классная строка с очень одухотворенным наполнением!
# Давай посчитаем сколько раз мы поржали (lol)
lol_count = some_strange_string_variable.count("lol")
print ("мы поржали раз:", lol_count)
shit_count = some_strange_string_variable.count('shit')
print ("Но теперь нам не до смеха, кол-во подходов у белого друга:", shit_count)
print ("Давай посмотрим с чего все это начиналось?!")
print ("Помыли ли мы руки перед едой?", some_strange_string_variable.startswith("Мыть руки"))
print ("Это мы сразу жрать начали?", some_strange_string_variable.startswith("Жрать"))
print ("Ну и как ты думаешь чем это все закончилось, Счастьем?", some_strange_string_variable.endswith("СЧАСТЬЕ?"))
print ("А на самом деле, наверное ты дико срал, да?", some_strange_string_variable.endswith("ДИКО СРАТЬ!!!!"))
print ("Выздоравливай, братан, и мой руки перед едой теперь!!")
print ("       ")
whaaat = "ghbdtn vtyz pjden Dbnfkbr!"
print ("Сегодня к нам в друзья добавился парень и написал нам в сообщении что-то непонятное:", whaaat)
print ("""Возможно он забыл переключить раскладку клавиатуры, 
 давайте мы ему поможем и заблокируем к херам, но все же выясним перед этим что он написал""")
# Составим таблицу перевода:
replacements= {
    'q': 'й', 'w': 'ц', 'e': 'у', 'r': 'к', 't': 'е', 'y': 'н', 'u': 'г', 'i': 'ш', 'o': 'щ', 'p': 'з', '[': 'х', ']': 'ъ', 
    'a': 'ф', 's': 'ы', 'd': 'в', 'f': 'а', 'g': 'п', 'h': 'р', 'j': 'о', 'k': 'л', 'l': 'д', ';': 'ж', '\'': 'э',
    'z': 'я', 'x': 'ч', 'c': 'с', 'v': 'м', 'b': 'и', 'n': 'т', 'm': 'ь', ',': 'б', '.': 'ю' # В данном случае '\'' - тут \ является экранированием специального символа '
              }
print ("Давай переведем что же написал нам этот парень:", whaaat.translate(str.maketrans(replacements)))
print ("Похоже мы забыли заглавные буквы в нашем словаре замен (replacements), давай добавим только 1 букву, остальные вы можете добавить сами")
print(type(replacements)) # https://www.w3schools.com/python/python_ref_dictionary.asp
replacements.update({'D': 'В'})
print ("Давай переведем что же написал нам этот парень:", whaaat.translate(str.maketrans(replacements)))
print ("Ура, получилось, теперь наша очередь ему писать письмо!")
our_letter = "Gjitk yf [eq Dbnfkbr, yt gbib c.lf ,jkmit, zcyj?"
print ("Наш ответ Виталику:", our_letter)

# Более подробно со строками вы можете ознакомиться на странице описания методов: https://www.w3schools.com/python/python_ref_string.asp
```

```python
###############################################################################################################
################################### Работа с датой и временем
###############################################################################################################
# Подробная документация: 
import datetime # импортируем пакет работы с датой и временем
import zoneinfo # Информация по временным зонам (часовым поясам)
import time

currDateTime = datetime.datetime.now() # https://docs.python.org/3/library/datetime.html
print(type(currDateTime))
print("Параметры пакета datetime:")
print("Минимальный год:", datetime.MINYEAR)
print("Максимальный год:", datetime.MAXYEAR)

print("     ")
print("Текущая дата и время:", currDateTime)
# Если нам нужна только дата, то
dateOnly = str(currDateTime.date()) # От даты и времени берем только дату
timeOnly = str(currDateTime.time()) # От даты и времени берем только время
print("Текущая дата:", dateOnly)
print("Текущее время:", timeOnly)

# А теперь давайте попробуем взять дату и время из прошлого
# class datetime.datetime(year, month, day, hour=0, minute=0, second=0, microsecond=0, tzinfo=None, *, fold=0)
print("     ")
zone = zoneinfo.ZoneInfo("Europe/Moscow")
print(type(zone))
birthDay = datetime.datetime(1990, 1, 16, 7, 0, 0, 0, zone)
print(type(birthDay))
print("ДР Сережи:", birthDay) # 1990-01-16 07:00:00+03:00 (+03:00 - часовой пояс Москвы) 
# в стандартной записи дата записывается так: год, месяц, день месяца
# Данный формат вывода даты соответствует такому шаблону времени в java: yyyy-MM-dd HH:mm:ssXXX
# Вывод ДР в формате ISO (стандарт)
print("ДР Сережи в формате ISO:", birthDay.isoformat())
print("ДР Сережи в своем формате:", birthDay.strftime("%d/%m/%Y %H:%M:%S"))
# И еще один вариант отформатировать мой ДР
formatted_bd = f"{birthDay.day:02d}|{birthDay.month:02d}|{birthDay.year} {birthDay.hour:02d}:{birthDay.minute:02d}:{birthDay.second:02d} {birthDay.tzinfo}"
print("ДР Сережи в своем формате (2):", formatted_bd)
# Еще один вариант использования
print("     ")
currTime = time.localtime() # https://docs.python.org/3/library/time.html#time.localtime
print(type(currTime))
print("Текущая дата, год:", currTime.tm_year)
print("Текущая дата, месяц:", currTime.tm_mon)
print("Текущая дата, день в месяце:", currTime.tm_mday)
print("Текущая дата, день в неделе:", currTime.tm_wday)
print("Текущая дата, день в году:", currTime.tm_yday)
print("----------------")
print("Текущее время, час:", currTime.tm_hour)
print("Текущее время, минута:", currTime.tm_min)
print("Текущее время, секунда:", currTime.tm_sec)
print("Текущее время, временная зона:", currTime.tm_zone)
```
